```python
import tensorflow as tf
print(tf.__version__) # check the version (should be 2.x+)

import numpy as np
import matplotlib.pyplot as plt

# Create features
X = tf.constant([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

# Create labels
y = tf.constant([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

# Visualize it
plt.scatter(X, y);
```

```python



# Set random seed
tf.random.set_seed(42)

# Create a model using the Sequential API
model = tf.keras.Sequential([
  tf.keras.layers.Dense(50, activation=None),
  tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(loss='mae', # mae is short for mean absolute error
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics=["mae"])

# Fit the model
# model.fit(X, y, epochs=5) # this will break with TensorFlow 2.7.0+
model.fit(tf.expand_dims(X, axis=-1), y, epochs=100) # start with this

# predict with 5 epochs
new_x = np.array([17.0, 23.0])

prediction = model.predict(new_x)

print(f"Prediction for {new_x[0]}: {prediction[0]}") # x1' = -11 - way off
print(f"Prediction for {new_x[1]}: {prediction[1]}") # x2' = -15

# as epochs increase
# e=5 , y=-11
# e=10, y=4.5
# e=50, y=11 (getting closer to 27)
# e=100, y=29.46 (passed 27)


```
